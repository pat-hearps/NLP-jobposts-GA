```python
cdict = {}
for did in dupIDs:
    ct = len(dfds[dfds.seekID == did])
#     print(ct)
    if ct in cdict.keys():
        cdict[ct] = cdict[ct] + 1
    else:
        cdict[ct] = 1
cdict
# just want to get an idea of how many repetitions there are of various IDs
```

```python
dudf = pd.DataFrame(columns=dfds.columns, index=dupIDs)
dudf.head()
```

```python
dfds.isnull().sum()
```

```python
def get_dup_nums(row):
    xdf = dfds[dfds.seekID == row.name]
    counts = [len(xdf[c].value_counts()) for c in xdf.columns]
    rdf = pd.Series(counts, index=xdf.columns)
    rdf.name = row.name
    return rdf
```

```python
%%time
dudf = dudf.apply(get_dup_nums, axis=1)
```

```python
dudf.head()
```

```python
dudf[dudf > 1].notnull().sum()
```

```python
prob_columns = ['company','salary','templatetext','title'] # only really care about differences in these fields
```

```python
%%time
##                     this is seeing how different the differences really are using fuzzywuzzy
check_dict = {}
for col in prob_columns:
    same = dfds[dfds.seekID.isin(dudf[dudf[col] > 1].index)]
    for IDcheck in same.seekID.unique():
#         print(IDcheck)
        for s in combinations([val for val in same[col][same.seekID == IDcheck]], 2):
            try:
                ratio = fuzz.partial_ratio(s[0],s[1])
#                 print(ratio)
                if ratio < 90:
                    check_dict[IDcheck] = s
            except:
                print(IDcheck, s)
                
# so now check_dict contains just those entries where the text is significantly different
```
